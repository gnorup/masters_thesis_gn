add feature: number of pauses that follow an article and precede content words (Vincze et al., 2022) to feature set


```python
# setup

import os
import sys
import pandas as pd

# add the root of the project to the path
sys.path.append("/Users/gilanorup/Desktop/Studium/MSc/MA/code/masters_thesis_gn/src")

from config.constants import GIT_DIRECTORY
from feature_extraction.features.article_pause_contentword import article_pause_contentword

task_name = "cookieTheft"

# load data
df = pd.read_csv(f"{GIT_DIRECTORY}/results/features/filtered/{task_name}_filtered.csv")

# path to data on server
word_timestamps = f"/Volumes/methlab/Students/Gila/word_timestamps/{task_name}/google/timestamps"
```

```python
# compute feature for each subject
article_pause_counts = []

subject_ids = df["Subject_ID"].dropna().unique()

for Subject_ID in subject_ids:
    print(f"adding features for subject {Subject_ID}")
    word_timestamps_path = os.path.join(word_timestamps, f"{Subject_ID}.csv")
    try:
        if os.path.exists(word_timestamps_path):
            count = article_pause_contentword(word_timestamps_path)
        else:
            print(f"file not found for {Subject_ID}")
            count = None
    except Exception as e:
        print(f"skipping {Subject_ID} due to error: {e}")
        count = None

    article_pause_counts.append(count)

df["article_pause_contentword"] = article_pause_counts

df.to_csv(f"{GIT_DIRECTORY}/results/features/filtered/{task_name}_filtered.csv", index=False)
print("new feature added and file saved")
```
