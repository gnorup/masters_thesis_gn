```python
# setup
import sys
import os
import pandas as pd

# add project root
sys.path.append("/Users/gilanorup/Desktop/Studium/MSc/MA/code/masters_thesis_gn/src")
from config.constants import GIT_DIRECTORY

# load data
cookieTheft = pd.read_csv(os.path.join(GIT_DIRECTORY, "results", "features", "filtered", "cookieTheft_filtered.csv"))
picnicScene = pd.read_csv(os.path.join(GIT_DIRECTORY, "results", "features", "filtered", "picnicScene_filtered.csv"))
journaling = pd.read_csv(os.path.join(GIT_DIRECTORY, "results", "features", "filtered", "journaling_filtered.csv"))
```

```python
# remove subjects where spontaneous speech task is empty

# cookieTheft
cookieTheft = cookieTheft[~cookieTheft["Subject_ID"].isin([138, 141, 469, 516, 1065])]

# picnicScene
picnicScene = picnicScene[picnicScene["Subject_ID"] != 390]

# journaling
journaling = journaling[~journaling["Subject_ID"].isin([474, 484])]
```

```python
# remove problematic features -> problem with definition or dictionary

cookieTheft = cookieTheft.drop(columns=["fam_content", "fam_nouns", "img_content", "img_nouns", "short_pause_count"])
picnicScene = picnicScene.drop(columns=["fam_content", "fam_nouns", "fam_verbs", "img_content", "img_nouns", "img_verbs", "short_pause_count"])
journaling = journaling.drop(columns=["fam_content", "fam_nouns", "img_content", "img_nouns", "short_pause_count"])
```

```python
# save back
cookieTheft.to_csv(os.path.join(GIT_DIRECTORY, "results", "features", "filtered", "cookieTheft_filtered.csv"), index=False)
picnicScene.to_csv(os.path.join(GIT_DIRECTORY, "results", "features", "filtered", "picnicScene_filtered.csv"), index=False)
journaling.to_csv(os.path.join(GIT_DIRECTORY, "results", "features", "filtered", "journaling_filtered.csv"), index=False)
```
