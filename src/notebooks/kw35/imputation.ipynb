impute certain feature-values for some subjects

```python
# setup
import os, sys
import pandas as pd

# project root
sys.path.append("/Users/gilanorup/Desktop/Studium/MSc/MA/code/masters_thesis_gn/src")

from config.constants import GIT_DIRECTORY

# load data
cookieTheft = pd.read_csv(os.path.join(GIT_DIRECTORY, "results", "features", "cookieTheft.csv"))
picnicScene = pd.read_csv(os.path.join(GIT_DIRECTORY, "results", "features", "picnicScene.csv"))
journaling = pd.read_csv(os.path.join(GIT_DIRECTORY, "results", "features", "journaling.csv"))
```

```python
# article_pause_contentword for subject 1370:

# impute from journaling-value and task-mean for cookieTheft and picnicScene
cT_mean = cookieTheft["article_pause_contentword"].mean()
pS_mean = picnicScene["article_pause_contentword"].mean()
feature_1370j = journaling.loc[journaling["Subject_ID"] == 1370, "article_pause_contentword"].iloc[0]
feature_1370cT = (cT_mean + feature_1370j) / 2
feature_1370pS = (pS_mean + feature_1370j) / 2
# save new values
cookieTheft.loc[cookieTheft["Subject_ID"] == 1370, "article_pause_contentword"] = feature_1370cT
cookieTheft.to_csv(os.path.join(GIT_DIRECTORY, "results", "features", "cookieTheft.csv"), index=False)
picnicScene.loc[picnicScene["Subject_ID"] == 1370, "article_pause_contentword"] = feature_1370pS
picnicScene.to_csv(os.path.join(GIT_DIRECTORY, "results", "features", "picnicScene.csv"), index=False)
```

```python
# mattr_30:

# cookieTheft: 389, 623, 957, 1155, 1276
cT_subjects = [389, 623, 957, 1155, 1276]
mean_cT = cookieTheft["mattr_30"].mean()
for subject in cT_subjects:
    feature_j = journaling.loc[journaling["Subject_ID"] == subject, "mattr_30"].iloc[0]
    feature_pS = picnicScene.loc[picnicScene["Subject_ID"] == subject, "mattr_30"].iloc[0]
    feature_cT = (((feature_pS + feature_j) / 2) + mean_cT) / 2
    cookieTheft.loc[cookieTheft["Subject_ID"] == subject, "mattr_30"] = feature_cT

# journaling: 1105
mean_j = journaling["mattr_30"].mean()
feature_1105cT = cookieTheft.loc[cookieTheft["Subject_ID"] == 1105, "mattr_30"].iloc[0]
feature_1105pS = picnicScene.loc[picnicScene["Subject_ID"] == 1105, "mattr_30"].iloc[0]
feature_1105j = (((feature_1105pS + feature_1105cT) / 2) + mean_j) / 2

# save new values
cookieTheft.to_csv(os.path.join(GIT_DIRECTORY, "results", "features", "cookieTheft.csv"), index=False)
journaling.loc[journaling["Subject_ID"] == 1105, "mattr_30"] = feature_1105j
journaling.to_csv(os.path.join(GIT_DIRECTORY, "results", "features", "journaling.csv"), index=False)
```
