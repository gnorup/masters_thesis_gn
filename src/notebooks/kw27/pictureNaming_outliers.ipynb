```python
# tried to remove picture-naming scores < 10 / impute them with mean but doesn't improve performance in model-comparison (makes it worse, actually) 

# setup
import sys
import os
import pandas as pd
import numpy as np

# add project root
sys.path.append("/Users/gilanorup/Desktop/Studium/MSc/MA/code/masters_thesis_gn/src")

from config.constants import GIT_DIRECTORY

# load data
scores_path = os.path.join(GIT_DIRECTORY, "data", "language_scores_all_subjects.csv")
df = pd.read_csv(scores_path)
```

```python
# subjects with PictureNaming-score below 10
outliers = df[df["PictureNamingScore"] < 10]
print(outliers[["Subject_ID", "PictureNamingScore"]])
print("Number of outliers:", outliers.shape[0])
```

```python
# remove those subjects to see if it changes results
df["PictureNamingScore"] = df["PictureNamingScore"].where(df["PictureNamingScore"] >= 10, np.nan)
save_path = os.path.join(GIT_DIRECTORY, "data", "language_scores_all_subjects_cleaned.csv")
df.to_csv(save_path, index=False)

print(f"saved cleaned file with {df.shape[0]} subjects to:\n{save_path}")
```

```python
scores_path = os.path.join(GIT_DIRECTORY, "data", "language_scores_all_subjects.csv")
df = pd.read_csv(scores_path)

# calculate mean (ignoring NaNs)
mean_score = df["PictureNamingScore"].mean()
print(f"Mean PictureNamingScore (excluding NaNs): {mean_score:.2f}")
```

```python
# try with imputation instead 
# create a copy to preserve original
df_imputed = df.copy()
# set low scores to NaN so they get imputed
df_imputed["PictureNamingScore"] = df_imputed["PictureNamingScore"].where(df_imputed["PictureNamingScore"] >= 10, np.nan)
# then impute
df_imputed["PictureNamingScore"] = df_imputed["PictureNamingScore"].fillna(mean_score)

# optional: save it
save_path = os.path.join(GIT_DIRECTORY, "data", "language_scores_all_subjects_imputed.csv")
df_imputed.to_csv(save_path, index=False)
```
