```python
import pandas as pd

# Define your paths
paths = {
    "1min": "/Users/gilanorup/Desktop/Studium/MSc/MA/code/masters_thesis_gn/results/features/picture_description_1min.csv",
    "2min": "/Users/gilanorup/Desktop/Studium/MSc/MA/code/masters_thesis_gn/results/features/picture_description_2min.csv",
    "full": "/Users/gilanorup/Desktop/Studium/MSc/MA/code/masters_thesis_gn/results/features/picture_description.csv"
}

for name, path in paths.items():
    df = pd.read_csv(path)

    # calculate mean and SD for n_words and duration_used_sec
    mean_n_words = df["n_words"].mean()
    sd_n_words = df["n_words"].std()

    mean_duration = df["duration_used_sec"].mean()
    sd_duration = df["duration_used_sec"].std()

    n_subjects = df["Subject_ID"].nunique()

    print(f"\n=== {name.upper()} ===")
    print(f"Number of subjects: {n_subjects}")
    print(f"Mean n_words: {mean_n_words:.2f} (SD = {sd_n_words:.2f})")
    print(f"Mean duration_used_sec: {mean_duration:.2f} (SD = {sd_duration:.2f})")
```
