```python
import pandas as pd

# load CSV
file_path = "/Users/gilanorup/Desktop/Studium/MSc/MA/dataset/all_subject_data.csv"
df = pd.read_csv(file_path)

# define age groups
bins = [50, 65, 75, 100]  # age ranges
labels = ["<65", "65–75", ">75"]
df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels, right=False)

# mean by Gender
gender_means = df.groupby("Gender")[["PictureNamingScore", "SemanticFluencyScore", "PhonemicFluencyScore"]].mean()
gender_sd = df.groupby("Gender")[["PictureNamingScore", "SemanticFluencyScore", "PhonemicFluencyScore"]].std()

# mean by Country
country_means = df.groupby("Country")[["PictureNamingScore", "SemanticFluencyScore", "PhonemicFluencyScore"]].mean()
country_sd = df.groupby("Country")[["PictureNamingScore", "SemanticFluencyScore", "PhonemicFluencyScore"]].std()

# mean by AgeGroup
agegroup_means = df.groupby("AgeGroup")[["PictureNamingScore", "SemanticFluencyScore", "PhonemicFluencyScore"]].mean()
agegroup_sd = df.groupby("AgeGroup")[["PictureNamingScore", "SemanticFluencyScore", "PhonemicFluencyScore"]].std()

print("\nMean scores by Gender:")
print(gender_means, gender_sd)

print("\nMean scores by Country:")
print(country_means, country_sd)

print("\nMean scores by Age Group:")
print(agegroup_means, agegroup_sd)
```
